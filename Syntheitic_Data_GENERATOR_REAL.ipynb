```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.decomposition import PCA
from sdv.single_table import CTGANSynthesizer
from sdv.metadata import SingleTableMetadata
```

```python
def synthetic(dff):
  df = pd.read_csv(dff)
  prompts = df['prompt_text'].tolist()
  model = SentenceTransformer("all-MiniLM-L6-v2") #Uses all-MiniLM-L6-v2 to transfrom sentence
  embeddings = model.encode(prompts, show_progress_bar=True) # Encodes prompts to embeddings
  pca = PCA(n_components=20)
  reduced_embeddings = pca.fit_transform(embeddings)
  reduced_embeddings_df = pd.DataFrame(reduced_embeddings)
  metadata = SingleTableMetadata()
  metadata.detect_from_dataframe(data=reduced_embeddings_df)
  ctgan = CTGANSynthesizer(metadata)
  ctgan.fit(reduced_embeddings_df)
  synthetic_data = ctgan.sample(1000)
  return synthetic_data
```

```python
data_leakage_prompts = "/content/drive/MyDrive/Colab Notebooks/data_leakage_prompts.csv"
jailbreak_prompts = "/content/drive/MyDrive/Colab Notebooks/jailbreak_prompts.csv"
phishing_prompts = "/content/drive/MyDrive/Colab Notebooks/phishing_prompts.csv"
prompt_injection_prompts = "/content/drive/MyDrive/Colab Notebooks/prompt_injection_prompts (1).csv"
```

```python
synthetic_data_leakage = synthetic(data_leakage_prompts)
np.save("Synthetic_Data_Leakage", synthetic_data_leakage)
```

```python
synthetic_jailbreak = synthetic(jailbreak_prompts)
np.save("Synthetic_Jailbreak", synthetic_jailbreak)
```

```python
synthetic_phishing = synthetic(phishing_prompts)
np.save("Synthetic_Phishing", synthetic_phishing)
```

```python
synethetic_prompt_injection = synthetic(prompt_injection_prompts)
np.save("Synthetic_Prompt_Injection", synethetic_prompt_injection)
```
