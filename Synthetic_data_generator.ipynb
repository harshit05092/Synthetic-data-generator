```python
import pandas as pd
```

```python
import numpy as np
from sentence_transformers import SentenceTransformer #Used to trasfrom sentences to embeddings
```

```python
df = pd.read_csv('/content/drive/MyDrive/synthetic_jailbreak_prompts.csv')
```

```python
df.head(5)
```

```python
df.value_counts('category')
```

```python
prompts = df['prompt'].tolist() #Converts prompts from dataframe column to list
```

```python
prompts
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2") #Uses all-MiniLM-L6-v2 to transfrom sentence
```

```python
embeddings = model.encode(prompts, show_progress_bar=True) # Encodes prompts to embeddings
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=20)
reduced_embeddings = pca.fit_transform(embeddings)
```

```python
np.save('embeddings.npy', reduced_embeddings)
```

```python
from sdv.single_table import CTGANSynthesizer
```

```python
from sdv.metadata import SingleTableMetadata
import pandas as pd

# Assuming 'reduced_embeddings' is your numpy array
# Convert the numpy array to a pandas DataFrame
reduced_embeddings_df = pd.DataFrame(reduced_embeddings)

metadata = SingleTableMetadata()

# Automatically detect metadata from the DataFrame
metadata.detect_from_dataframe(data=reduced_embeddings_df)

# You can view the detected metadata
print(metadata)
```

```python
from sdv.single_table import CTGANSynthesizer
import pandas as pd

# Assuming 'embeddings' is your numpy array
# Convert the numpy array to a pandas DataFrame
embeddings_df = pd.DataFrame(reduced_embeddings)

# Re-initialize CTGANSynthesizer with the correct metadata
ctgan = CTGANSynthesizer(metadata)
ctgan.fit(embeddings_df)  # Fit with the DataFrame

# Generate synthetic embeddings
synthetic = ctgan.sample(500)
```
